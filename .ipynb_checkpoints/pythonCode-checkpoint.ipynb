```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("amazon_sales.csv")
```

```python
# VÃ©rifier le type des colonnes et les valeurs manquantes
print(df.info())
```

```python
# Statistiques descriptives rapides
print(df.describe())
```

```python
# Supprimer le symbole â‚¹ et les virgules, puis convertir en float
df['discounted_price'] = df['discounted_price'].str.replace('â‚¹','').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('â‚¹','').str.replace(',','').astype(float)

# VÃ©rifier
print(df[['discounted_price','actual_price']].head())
```

```python
# Supprimer le % et convertir en float
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)

# VÃ©rifier le rÃ©sultat
print(df['discount_percentage'].head())
```

```python
# Remplacer les valeurs non convertibles en float par NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# VÃ©rifier les lignes oÃ¹ rating est NaN
print(df[df['rating'].isna()])
```

```python
df = df.dropna(subset=['rating'])
```

```python
# Convertir en float
df['rating'] = df['rating'].astype(float)

# VÃ©rifier
print(df['rating'].head())
```

```python
# Supprimer les virgules
df['rating_count'] = df['rating_count'].str.replace(',','')

# Convertir en int, remplacer les valeurs invalides par NaN
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# VÃ©rifier les valeurs NaN
print(df[df['rating_count'].isna()])
```

```python
df = df.dropna(subset=['rating_count'])
```

```python
print(df[['rating','rating_count']].head())
print(df.dtypes)
```

```python
# Moyenne du prix rÃ©el et du prix rÃ©duit par catÃ©gorie
category_price = df.groupby('category')[['actual_price','discounted_price']].mean().sort_values('actual_price', ascending=False)
print(category_price.head(10))
```

```python
category_discount = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
print(category_discount.head(10))
```

```python
top_rated_products = df.groupby('product_name')['rating'].mean().sort_values(ascending=False).head(10)
print(top_rated_products)
```

```python
# Extraire la premiÃ¨re catÃ©gorie avant |
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])

# Moyenne du prix rÃ©el par catÃ©gorie principale
main_category_price = df.groupby('main_category')['actual_price'].mean().sort_values(ascending=False).head(10)

# Barplot
plt.figure(figsize=(10,6))
sns.barplot(x=main_category_price.index, y=main_category_price.values)
plt.xticks(rotation=45)
plt.title('Top 10 catÃ©gories principales par prix moyen')
plt.show()
```

```python
# Top 10 produits les mieux notÃ©s
plt.figure(figsize=(10,6))
sns.barplot(x=top_rated_products.index, y=top_rated_products.values)
plt.xticks(rotation=90)
plt.title('Top 10 produits par note moyenne')
plt.show()
```

```python
# CrÃ©er la connexion Ã  la base
conn = sqlite3.connect("amazon_sales.db")
cursor = conn.cursor()

# Exporter le DataFrame nettoyÃ© dans SQLite
df.to_sql('sales', conn, if_exists='replace', index=False)

# VÃ©rifier que la table a Ã©tÃ© crÃ©Ã©e
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
# Top 10 produits par nombre dâ€™avis

query = """
SELECT product_name, SUM(rating_count) as total_reviews
FROM sales
GROUP BY product_name
ORDER BY total_reviews DESC
LIMIT 10;
"""
cursor.execute(query)
results = cursor.fetchall()

# Dans un DataFrame
top_products = pd.DataFrame(results, columns=['product_name', 'total_reviews'])

# Tronquer les noms de produits Ã  50 caractÃ¨res pour la lisibilitÃ©
top_products['product_name'] = top_products['product_name'].apply(lambda x: x[:50] + '...' if len(x) > 50 else x)

top_products
```

```python
# Si la colonne main_category n'existe pas encore dans le DataFrame, on la crÃ©e
if 'main_category' not in df.columns:
    df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])
    # Mettre Ã  jour la table SQL
    df.to_sql('sales', conn, if_exists='replace', index=False)

# Top 10 catÃ©gories principales par prix moyen
query = """
SELECT main_category, AVG(actual_price) as avg_price
FROM sales
GROUP BY main_category
ORDER BY avg_price DESC
LIMIT 10;
"""
cursor.execute(query)
results = cursor.fetchall()

# Mettre le rÃ©sultat dans un DataFrame pour plus de lisibilitÃ©
top_categories = pd.DataFrame(results, columns=['main_category', 'avg_price'])

# Afficher le tableau
top_categories
```

```python
# Produits avec rating >= 4 et discount >= 50%
query = """
SELECT product_name, rating, discount_percentage
FROM sales
WHERE rating >= 4 AND discount_percentage >= 50
ORDER BY rating DESC, discount_percentage DESC;
"""
cursor.execute(query)
results = cursor.fetchall()

# Mettre le rÃ©sultat dans un DataFrame pour lisibilitÃ©
top_discounted_high_rating = pd.DataFrame(results, columns=['product_name', 'rating', 'discount_percentage'])

# Tronquer les noms de produits longs pour plus de lisibilitÃ©
top_discounted_high_rating['product_name'] = top_discounted_high_rating['product_name'].apply(lambda x: x[:50] + '...' if len(x) > 50 else x)

# Afficher le tableau
top_discounted_high_rating
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Barplot
plt.figure(figsize=(10,6))
sns.barplot(x='total_reviews', y='product_name', data=top_products)
plt.title('Top 10 produits par nombre d\'avis')
plt.xlabel('Nombre total d\'avis')
plt.ylabel('Produit')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='avg_price', y='main_category', data=top_categories)
plt.title('Top 10 catÃ©gories principales par prix moyen')
plt.xlabel('Prix moyen')
plt.ylabel('CatÃ©gorie principale')
plt.show()
```

```python
# pour Excel

top_products.to_excel("top_products.xlsx", index=False)

top_categories.to_excel("top_categories.xlsx", index=False)

top_discounted_high_rating.to_excel("top_discounted_high_rating.xlsx", index=False)
```
